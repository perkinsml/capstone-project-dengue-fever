# Data

The environmental data (i.e. the features) and dengue fever data (i.e. the outcome) used for this project have been sourced from the [National Oceanic Atmospheric Administration's Dengue Forecasting project website](http://dengueforecasting.noaa.gov/).  

The features and outcomes/labels data have been downloaded from this site in a series of csv and excel files.

The dataset includes weekly dengue fever counts for the two cities - the San Juan-Carolina-Caguas Metropolitan Statistical Area, hereafter referred to as San Juan, Puerto Rico and Iquitos, Peru - for the following time periods:
- <b>San Juan, Puerto Rico</b>: between the weeks commencing April 30th (week 18) 1990 and April 23rd (week 17) 2009 i.e. 19 years of weekly data = 988 data points
- <b>Iquitos, Peru</b>: between the weeks commencing July 1st (week 27) 2000 and June 25th (week 26) 2009 i.e. 9 years of weekly data = 468 data points

The feature data includes a mix of weekly and daily measurements over varying time periods, with some daily data stretching as far back as 1956.

## Outcome data

The outcome data is stored in two csv files (1 for each city) and includes the following fields:
<ol>
    <li> <b>season:</b> the dengue transmission season</li>
<li> <b>season_week:</b> week of the season (not the calendar year - the season starts following the week with historically lowest dengue cases over the years 1990-2009 and 2000-2009 for San Juan and Iquitos respectively)</li>
<li> <b>week_start_date:</b> date of the first day of the week.  Note, each year starts on Jan 1st and there are 52 weeks in each year, so week 52 of each year may contain an additional day (or two, in leap years). </li>
<li> <b>denv1_cases:</b> number of laboratory confirmed cases with DENV1</li>
<li> <b>denv2_cases:</b> number of laboratory confirmed cases with DENV2</li>
<li> <b>denv3_cases:</b> number of laboratory confirmed cases with DENV3</li>
<li> <b>denv4_cases:</b> number of laboratory confirmed cases with DENV4</li>
<li> <b>other_positive_cases:</b> laboratory-positive cases without type identified</li>
<li> <b>additional_cases (for San Juan only):</b> at times not all specimens submitted were tested due to overload of the capacity for testing or incomplete case information. For those weeks, the number of additional laboratory-positive cases among those not tested was estimated by multiplying the number of untested cases by the rate of laboratory-positive cases amongst those that were tested</li>
<li> <b>total_cases:</b> the sum of all cases, <b><u>the target variable for modeling</u></b></li>
</ol>

## Features data

The environmental data included in this study are from a variety of sources, including 
ground observations from weather stations, remote sensing data from satellites and reanalysis data from climate models.  Each data source has different limitations and quality issues:

- <b>Ground observations</b> are generally the optimal representation of actual local conditions
- <b>Remotely sensed observations</b> are generally an excellent representation of precipitation and vegetation conditions for a location
- <b>Reanalysis data</b> are a good representation of a given area, especially when other data sources are not available

The environmental data for this study are at varying spatial scales.  Station data 
provide point-based estimates of the local climate conditions, whereas the reanalysis and 
remotely sensed data provide data based on grid cells of varying resolutions.

### Overview
The features data are stored in a number of csv and excel files and are provided at both weekly and daily frequencies, depending on the feature.  

Each csv and excel file includes the year, month and week or day number (depending on the feature) in the year, alongside the features data outlined in the sections below.  

<b>Helpful abbreviations</b>
- GHCN: Global Historical Climate Network
- NCEP: National Centres for Environmental Prediction
- NDVI: Normalised Differnce Vegetation Index
- NOAA: National Oceanic and Atmospheric Administration
- PERSIANN: Precipitation Estimation from Remotely Sensed Information using Artificial Neural Networks

#### PERSIANN satellite precipitation data

- <b>precipitation_amt_mm:</b> total weekly precipitation in millimetres
    
Missing values are listed as -9999.

#### NOAA's NCEP reanalysis data
NOAA's NCEP climate forecast system reanalysis data includes the daily, modeled estimates of the fields 
listed below on a 0.5x0.5 degree scale.  These estimates are derived from a climate model representing the global interaction between Earth's oceans, land and atmosphere. This data includes variables that are not available from other source (e.g. humidity) and can be used to complement missing observational data.
- <b>reanalysis_air_temp_k:</b> mean air temperature, in degrees Kelvin
- <b>reanalysis_avg_temp_k:</b> average air temperature, in degrees Kelvin
- <b>reanalysis_dew_point_temp_k:</b> mean dew point temperature, in degrees Kelvin
- <b>reanalysis_max_air_temp_k:</b> maximum air temperature, in degrees Kelvin
- <b>reanalysis_min_air_temp_k:</b> minimum air temperature, in degrees Kelvin
- <b>reanalysis_precip_amt_kg_per_m2:</b> total precipitation, in kg per m2
- <b>reanalysis_relative_humidity_percent:</b> mean relative humidity percentage
- <b>reanalysis_sat_precip_amt_mm:</b> total precipitation, in millimetes
- <b>reanalysis_specific_humidity_g_per_kg:</b> mean specific humidity, in g per kg
- <b>reanalysis_tdtr_k:</b> diurnal temperature range, in degrees Kelvin

#### NOAA's NDVI data
The NDVI is the basic index for measuring the 'greeness' of the earth's surface.  This 
index provides an estimation of the density of surface vegetation and is derived from 
surface reflection data captured from NOAA weather satelites.

NDVI values range from -0.1 to 0.703 and are unitless.  Values greater than 0.1 generally 
denote increasing degrees of 'greeness'and intensity of vegetation.  Values between 0 and 0.1 
are characteristic of rocks and soil, while values less than 0 sometimes indicate 
cloud, rain or snow.

NOAA's NDVI data includes the following fields and weekly measurements on a 0.5x0.5 degree scale,
for the 4 pixels closest to the city centroids for both San Juan and Iquitos:
- <b>ndvi_ne:</b> pixel northeast of city centroid
- <b>ndvi_nw:</b> pixel northwest of city centroid
- <b>ndvi_se:</b> pixel southeast of city centroid
- <b>ndvi_sw:</b> pixel southwest of city centroid
    
These data are available on a weekly basis from 1981.


###  Features data strategy
Given that the climate station measurements are the best representation of local conditions, my planned approach is to use these station measurements wherever possible as my feature data, and to supplement these with reanalysis data if/where required. 

I'll likely use the PERSIANN satellite precipitation estimates, as well as the NDVI data.  There are also be some interesting features (e.g. relative humidity) in the reanalysis data that are not available from the other data sources.

Throughout this analysis, I plan to compare the correlations between my features and the outcome variable over lagged timeframes (e.g the incidence of dengue fever may be related to the temperature 12 weeks prior).  For this reason, I'll create a features dataset that starts prior to the beginning of the outcome data to enable me to perform this analysis.  Features data will be created for the following timeframes:
- <b>San Juan:</b> Jan 1st, 1989 - April 29th, 2009 vs. the outcome data timeframe of April 30th, 1990 - April 29th, 2009
- <b>Iquitos:</b> Jan 1st, 1999 - July 1st, 2009 vs. the outcome data timeframe of July 1st, 2000 - July 1st, 2009

All features data will need to be aggregated on a weekly basis to enable it to used as a predictor for the weekly dengue outcome data.

```python
# Define function to create and insert datetime object into first column 
# of dataframe, using existing fields within the dataframe and return it 
# sliced between two dates passed to the function as strings

def date_and_slice(df, start_date, end_date):
    '''This function takes a dataframe, and a string start_date and end_date
       It will insert a datetime object into the dataframe, derived from existing
       fields within the dataframe, subset it between the start_date and end_date
       string arguments, and return the dataframe with a reset index.'''
    # Insert DATE string into index 0
    df.insert(0, 'DATE', df.YYYY.astype('str') + \
              '/' + df.MM.astype('str') + \
              '/' + df.DD.astype('str'))

    # Convert DATE string to datetime object 
    df.DATE = pd.to_datetime(df.DATE)
    
    # Return subset of dataframe between two dates with the index reset
    mask = (df.DATE >= start_date) & (df.DATE <= end_date)

    return df.loc[mask,:].reset_index(drop=True)
```

```python
# Define function to return week of year, based on the DD field
# In line with outcome data, all days falling after week 51 will be assigned to week 52
def weekofyear(day_of_year):
    '''This function accepts a day number within the year and returns the corresponding week number.'''
    if day_of_year > 364:
        return 52
    else:
        divided_by_seven = day_of_year / 7
        return math.ceil(divided_by_seven)
```

```python
# Insert week of year column, as calculcated by weekofyear function. This will be used for calculating averages
sj_features_df.insert(4, 'WEEK_OF_YEAR', sj_features_df.DD.map(weekofyear))

# Insert week of year column, as derived from date object.  This may be useful later.
sj_features_df.insert(5, 'CAL_WEEK_OF_YEAR', sj_features_df.DATE.dt.weekofyear)
sj_features_df.head()
```

```python
# Define function to aggregate a df by a specified function on a specified field
# The function will return the aggregated df, with the index reset
def weekly_calc(df, field, agg):
    '''This function accepts a df, the name of a field in the df (as a string)
    and a string that specifies the aggregation required (e.g. max, min or mean).
    The function returns the aggregated data only.'''

    if agg == 'max':
        weekly_agg = pd.DataFrame(df.groupby(['YYYY','WEEK_OF_YEAR'])[field].max())
    elif agg == 'min':
        weekly_agg = pd.DataFrame(df.groupby(['YYYY','WEEK_OF_YEAR'])[field].min())  
    elif agg == 'mean':
        weekly_agg = pd.DataFrame(df.groupby(['YYYY','WEEK_OF_YEAR'])[field].mean())
    else:
        print('Invalid entry')
        return
    
    weekly_agg.reset_index(inplace=True)
    return weekly_agg.iloc[:,2]
```
