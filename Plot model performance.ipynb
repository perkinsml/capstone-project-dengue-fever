```python
mods = ['ENet*', 'RFR', 'ETR', 'SVM', 'KN']
R2 = [0.544, 0.453, 0.395, 0.543, 0.361]
r2 = pd.DataFrame({'Models':mods, 'R2 scores':R2})
r2.set_index('Models', inplace=True)
```

```python
def plot_r2(df, city):
       
    plt.figure(figsize=(6, 5))
    
    # Define color palette
    pal = ['crimson', 'darkorange', 'c', 'b', 'm']
    
    # Plot seasonal means as a bar chart
    plt.bar(r2.index, r2['R2 scores'], color=pal, width=0.5)
    
    plt.ylabel('R2 scores')
    plt.ylim(0, 1)
    plt.xlabel('Models')
    plt.xticks()
    #plt.xlim(0,season_totals_df.index.size)
    plt.title(city + ': Maximum R2 scores, per model with a 4 week lead time');
```

```python
plot_r2(r2, 'San Juan')
```
